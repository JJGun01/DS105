#  P1: Loving Food

```python
# Add your web scraping code here
url_general_page = 'https://www.bbcgoodfood.com/search?tab=recipe&diet=vegetarian&sort=rating&q=pie'
response = requests.get(url_general_page)
soup = BeautifulSoup(response.text)
a = soup.find_all('div', attrs={'class':'card__section card__content'})
```

```python
#Python way
link =[]
name =[]
calories=[]
for entry in a:
    t = 'https://www.bbcgoodfood.com' + entry.find('a').get('href')
    if entry.find('a').get('href')[1] =='r' and entry.find('a').get('href')[4] == 'i' and entry.find('a').get('href')[7]=='s':
        link.append(t)
    one_name = str(entry.find('h2'))
    one_name = one_name.split('inherit">')[-1].replace('</h2>','').replace(' &amp;', ' and').replace('','')
    if one_name != 'None':
        name.append(one_name)


parsed_data = {"name":name, "link":link}
df = pd.DataFrame(parsed_data)
df

```

```python
#pandas way
link=[]
name=[]

def function(a):
    t = 'https://www.bbcgoodfood.com' + a.find('a').get('href')
    if a.find('a').get('href')[1] =='r' and a.find('a').get('href')[4] == 'i' and a.find('a').get('href')[7]=='s':
        link.append(t)
    one_name = str(a.find('h2'))
    one_name = one_name.split('inherit">')[-1].replace('</h2>','').replace(' &amp;', ' and').replace('','')
    if one_name != 'None':
        name.append(one_name)
    parsed_data = {"name":name, "link":link}
    df = pd.DataFrame(parsed_data)
    return df


df_civica = pd.concat([function(card) for card in a]).drop_duplicates()
df_civica
```

```python
calories_list= []
carbs_list=[]
author_list =[]
for individual in link:
    url_specific_page = str(individual)
    response = requests.get(url_specific_page)
    soup = BeautifulSoup(response.text)
    x = soup.find_all('td', {'class', 'key-value-blocks__value'})
#0 and 3 correspond to the index number of calories and carbs
    calories = str(x[0])
    calories = calories.split('value">')[1].replace('</td>','').replace('<!-- -->g</td>','').replace('<!-- -->g','')
    calories_list.append(calories)
    carbs = str(x[3])
    carbs = carbs.split('value">')[1].replace('</td>','').replace('<!-- -->g</td>','').replace('<!-- -->g','')
    carbs_list.append(carbs)
    author = soup.find('a', {'class','link link--styled'})
    author = str(author)
    author = author.split('"author">')[1].replace(' – Cookery writer</a>','').replace('</a>','')
    author_list.append(author)
calories_list
carbs_list
author_list

df_1 = pd.DataFrame({'n_cal':calories_list,'n_carb':carbs_list,'author':author_list})
table = pd.concat([df,df_1], axis = 1)
table
```

# P2: Know your heroes 

```python
# Add your web scraping code here
author_list=[]
author_link_list=[]
author_para_list=[]
for individual in link:
    url_specific_page = str(individual)
    response = requests.get(url_specific_page)
    soup = BeautifulSoup(response.text, 'html.parser')
    author = soup.find('a', {'class','link link--styled'})
    author = str(author)
    author = author.split('"author">')[1].replace(' – Cookery writer</a>','').replace('</a>','')
    author_list.append(author)
    
    author_link = str(soup.find('a', {'class','link link--styled'}))
    author_link= 'https://www.bbcgoodfood.com' + author_link.split('href="')[1].split('" rel="author"')[0]
    author_link_list.append(author_link)

    url_specific_person_page = str(author_link)
    response = requests.get(url_specific_person_page)
    soup = BeautifulSoup(response.text,'html.parser')
    author_para =str(soup.find_all('p',{'author-archive-header__bio author-archive-header__bio--left'}))
    author_para = author_para.split('\r\n\r\n')[0].replace('&amp','')
    if len(author_para.split('\n                        ')) >= 2:
        author_para = author_para.split('\n                        ')[1].replace('\n                    </p>]','')
        author_para_list.append(author_para)
    else:
        author_para = None
        author_para_list.append(author_para)
    

df_2 = pd.DataFrame({'author': author_list,'bios':author_para_list ,'links': author_link_list})
df_2.to_json(r'/home/jiajian/author_info.json')
```
